```python
from PIL import Image
import glob
import os
```

```python
base_path = "D:/Documents/13_Projet/dlcv4ldb/02_Donnees"
empty_maps_path = "empty_maps/incarnam"
bg_folder_path = f"{base_path}/{empty_maps_path}"
print(f"{bg_folder_path=}")
```

```python
left_off, top_off, right_off, bot_off = (355, 10, 355, 135)
area_list = glob.glob(f"{bg_folder_path}/*")
for area_path in area_list:
    print(f"{area_path=}")
    area_full_list = glob.glob(f"{area_path}/full/*")
    for area_full_path in area_full_list:
        print(f"{area_full_path=}")
        img = Image.open(area_full_path, 'r')
        box_crop = (left_off, top_off, img.size[0] - right_off, img.size[1] - bot_off)
        print(f"{box_crop=}")
        img_crop = img.crop(box_crop)
        area_crop_path = area_full_path.replace("full", "crop")
        img_crop.save(area_crop_path)
```
